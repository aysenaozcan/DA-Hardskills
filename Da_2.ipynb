```python
 #Fonksiyon ile faktoriyel hesaplama
def faktoriyel(sayi):
  fakto=1
  if(sayi==1 or sayi==0):
    print("Faktoriyeli= ", fakto)
  else:
    while (sayi>=1):
      fakto*=sayi
      sayi-=1
      print("Faktoriyeli",fakto)
      print(f"Faktoriyeli={fakto}")

faktoriyel(6)

```

```python
def asal_mi(sayi):
  if(sayi==1):
    return False
  elif(sayi==2):
    return True
  else:
    for i in range(2,sayi):
      if(sayi%i==0):
        return False
    return True
```

```python
while True:
  sayi=input("Sayı:")
  if(sayi=="q"):
    break
  else:
    sayi=int(sayi)
    if (asal_mi(sayi)):
      print("Sayi asaldir")
    else:
      print("Sayi asal degildir")
```
