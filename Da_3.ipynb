```python
class Employee():
  def __init__(self,name,surname,salary):
    print("Employee sınıfının init fonksiyonu çağrıldı.")
    self.name=name
    self.surname=surname
    self.salary=salary
  def show_the_data(self):
    print(f"Name: {self.name}, Surname: {self.surname}, Salary: {self.salary}")
    print("Employee sınıfının bilgileri gösterildi.")
  def dep_change(self,new_dep):
    print("Employee sınıfının departmanı değiştirildi.")
    self.dep=new_dep
```

```python
class Manager(Employee): #inheritance
  def raise_salary(self,raise_amount):
    print("Maaş artıyor.")
    self.salary+=raise_amount
```

```python
class Employee():
  def __init__(self,name,surname,salary):
    print("Employee sınıfının init fonksiyonu çağrıldı.")
    self.name=name
    self.surname=surname
    self.salary=salary
  def show_the_data(self):
    print(f"Name: {self.name}, Surname: {self.surname}, Salary: {self.salary}")
    print("Employee sınıfının bilgileri gösterildi.")
  def dep_change(self,new_dep):
    print("Employee sınıfının departmanı değiştirildi.")
    self.dep=new_dep
```

```python
class Manager(Employee): #overriding
  def __init__(self,name, surname, salary, dep, number_of_emp):
    super().__init__(name, surname, salary) # Call parent's __init__
    self.dep=dep
    self.number_of_emp=number_of_emp
    print("Manager sınıfının init fonksiyonu çağrıldı.")
  def raise_salary(self,raise_amount):
    self.salary+=raise_amount
    print("Manager sınıfının init fonksiyonu çağrıldı.")
```

```python
class Manager(Employee):

    def __init__(self, name, surname, salary, dep, number_of_emp):
        super().__init__(name, surname, salary)

        print("Yonetici sınıfının initi çağırıldı.")
        self.dep = dep
        self.number_of_emp = number_of_emp

    def show_the_data(self):
        print("Calisan sınıfının bilgileri gosterildi.")
        print(f"Isim: {self.name}, Soyisim: {self.surname}, Maas: {self.salary}, Dep: {self.dep}, Kisi_Sayisi: {self.number_of_emp}")

    def raise_salary(self, raise_amount):
        print("Maaş arttı.")
        self.salary+=raise_amount
```

```python
a=Manager("Mustafa", "Yilmaz", 20000,"Arge",10)
```

```python
a.show_the_data()
```

```python
a.raise_salary(5000)
print("Maaş artıyor.")
```
